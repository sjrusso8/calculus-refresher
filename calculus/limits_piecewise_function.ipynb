```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
from IPython.display import display,Math
```

```python
# Using sym to declare a variable
x = sym.symbols('x')

# the function
fx = x**3

# limit
lim_pnt = 1.5
lim = sym.limit(fx,x,lim_pnt)

# display the function
display(Math('\\lim_{x\\to %g} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

```python
# Create the function using lambda
fxx = sym.lambdify(x,fx)
xx = np.linspace(-5, 5, 200)

plt.plot(xx, fxx(xx))
plt.plot(lim_pnt,lim, 'ro')

plt.show()
```

```python
# Define a function
fx = (x ** 2)/(x-2)

# make the function a lambda
fxx = sym.lambdify(x,fx)

# showing that the function goes to infinit cause of the (x-2) in the function above
lim_pnt = 2
lim = sym.limit(fx, x, lim_pnt)
lim
```

```python
#  Creating a plot for that function above

# create a x axis as input to the function
xx = np.linspace(1,3, 200)

plt.plot(xx, fxx(xx))
display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

## Exercise: plot more complex functions & compute some limits

### Plot more complex functions

```python
# Creating the variables and the functions
x = sym.symbols('x')

fx = sym.sqrt(x+1)*sym.exp(-x)
gx = sym.cos(x+ sym.sin(x))

fxx = sym.lambdify(x, fx)
gxx = sym.lambdify(x, gx)
```

```python
# Plotting the stuff
xx = np.linspace(0,10,100)

plt.plot(xx, fxx(xx), label="f(x)")
plt.plot(xx, gxx(xx), label="g(x)")

plt.legend()
plt.show()
```

### Limit of two functions

```python
# Finding that a two functions to a limit is the same

lim_pnt = 5

lim_fx = sym.limit(fx,x,lim_pnt)
lim_gx = sym.limit(gx,x,lim_pnt)

# compute limit of f(x)/g(x)
hx = fx/gx
lim_fgx = sym.limit(hx, x, lim_pnt)

display(Math('\\frac{\\lim_{x \\to %g} f(x)}{\\lim_{x \\to %g} g(x)} = \\frac{%g}{%g} = %g' \
     %(lim_pnt, lim_pnt, lim_fx, lim_gx, lim_fx/lim_gx)))


display(Math('\\lim_{x\\to %g} \\frac{f(x)}{g(x)} = %g' %(lim_pnt, lim_fgx)))
```

## SECTION: Piecewise Functions

```python
from sympy.abc import x

piece1 = 0
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise( (piece1, x<0), (piece2, (x>=0) & (x<3)), (piece3, x>=3) )
```

```python
fx
```

```python
fxx = sym.lambdify(x, fx)

xx = np.linspace(-10, 5, 1234)

plt.plot(xx, fxx(xx))
```
