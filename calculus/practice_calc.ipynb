```python
x = sym.symbols('x')

fx = (6*x**2 + 7*x)**4
fx
```

```python
sym.diff(fx)
```

```python
xx = np.linspace(-10,10,100)

fx_lambda = sym.lambdify(x, fx)(xx)
dfx_lambda = sym.lambdify(x, sym.diff(fx))(xx)
```

```python
# plot it!
plt.plot(xx,fx_lambda,'k',linewidth=2)
plt.plot(xx,dfx_lambda,'r',linewidth=2)
plt.show()
```

## Graphing a fun one

```python
fx = (sym.sin(sym.pi*x)/(x**2 - 1))
gx = x**2 - 1
hx = sym.sin(sym.pi*x)
```

```python
xx = np.linspace(-5,5,1000)

fx_lambda = sym.lambdify(x, fx)(xx)
dfx_lambda = sym.lambdify(x, sym.diff(fx))(xx)
gx_lambda = sym.lambdify(x, gx)(xx)
hx_lambda = sym.lambdify(x, hx)(xx)

plt.plot(xx,fx_lambda,'k',linewidth=2)
plt.plot(xx,gx_lambda,'r',linewidth=2)
plt.plot(xx,hx_lambda,'g',linewidth=2)
plt.plot(xx,hx_lambda,'g',linewidth=2)
plt.ylim(-3,3)
plt.xlim(-3,3)
plt.grid(True, which='both')
plt.show()
```

## Taylor Series


Having the higher order derivatives match a function you are trying to emulate

```python
cosx = sym.cos(x)

# Adding a polynoial to estimate that same graph
#   c + c1*x c2*X**2
tx = 1 + 0*x + (-1/2)*x**2
gx = 1 + 0*x + (-1/2)*x**2 + 0*x**3 + (1/24)*x**4

xx = np.linspace(-5,5,1000)

cosx_lambda = sym.lambdify(x, cosx)(xx)
tx_lambda = sym.lambdify(x, tx)(xx)
gx_lambda = sym.lambdify(x, gx)(xx)

plt.plot(xx,cosx_lambda,'k',linewidth=2, label="Cos(x)")
plt.plot(xx,tx_lambda,'g',linewidth=2, label="1 + -1/2*x**2")
plt.plot(xx,gx_lambda,'b',linewidth=2, label="1 + -1/2*x**2+(1/24)**4")
plt.ylim(-1.5,1.5)
plt.xlim(-5,5)
plt.legend()
plt.show()
```
