```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
from IPython.display import display,Math
from matplotlib.patches import Polygon
```

## Simple Chain Run Stuff
### Also plotting


```python
x = sym.symbols('x')

fx = (6*x**2 + 7*x)**4
fx
```

```python
sym.diff(fx)
```

```python
xx = np.linspace(-10,10,100)

fx_lambda = sym.lambdify(x, fx)(xx)
dfx_lambda = sym.lambdify(x, sym.diff(fx))(xx)
```

```python
# plot it!
plt.plot(xx,fx_lambda,'k',linewidth=2)
plt.plot(xx,dfx_lambda,'r',linewidth=2)
plt.show()
```

```python
gx = 2 ** (1 - 6*x)

gx_lambda = sym.lambdify(x, gx)(xx)
dgx_lambda = sym.lambdify(x, sym.diff(gx))(xx)
```

```python
# plot it!
plt.plot(xx,gx_lambda,'k',linewidth=2)
plt.plot(xx,dgx_lambda,'r',linewidth=2)
plt.show()
```

```python
yx = sym.sin(3*x) / (1 + x ** 2)
dyx = sym.diff(yx)

yx_lambda = sym.lambdify(x, yx)(xx)
dyx_lambda = sym.lambdify(x, sym.diff(dyx))(xx)
```

```python
plt.plot(xx,yx_lambda,'k',linewidth=2)
plt.plot(xx,dyx_lambda,'r',linewidth=2)
plt.show()
```
