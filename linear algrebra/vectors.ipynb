```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display,Math
```

## Vectors in Numpy

```python
row_vec = np.array([4,2])
col_vec = np.array([ [-2], [3]])
```

```python
display(Math(sym.latex(sym.sympify(row_vec))))
display(Math(sym.latex(sym.sympify(col_vec))))
```

```python
# plotting vectors in the standard position
plt.plot([0,row_vec[0]], [0, row_vec[1]], 'r', label="row vector")
plt.plot([0,col_vec[0]], [0, col_vec[1]], 'b', label="column vector")

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()

```

## Simple exercise for vectors

```python
vec1 = np.array([3,0,-4])
vec2 = np.array([-1,1,3])

display(Math(sym.latex(sym.sympify(vec1))))
display(Math(sym.latex(sym.sympify(vec2))))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=plt.figaspect(1))
# This is important
ax = fig.add_subplot(111, projection='3d')

# draw vectors
ax.plot([0,vec1[0]],[0,vec1[1]],[0,vec1[2]],'b',linewidth=3)
ax.plot([0,vec2[0]],[0,vec2[1]],[0,vec2[2]],'r',linewidth=3)

# guidelines
ax.plot([-5,5],[0,0],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[-5,5],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[0,0],[-5,5],'--',color=[.7,.7,.7])

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)
plt.show()
```

## Vector Scaler 

```python
v = np.array([.5,1])
s = [1, -.5, 2, .5] 

for si in s:
    sv = si * v
    plt.plot([0,sv[0]], [0, sv[1]], "o-",linewidth=3,label='$\\lambda=%g$' %(si))
    
plt.axis('square')
plt.legend()
plt.axis([-3,3,-3,3])
plt.grid()
plt.show()
```

```python
#  Adding vectors

v1 = np.array([-1, 2])
v2 = np.array([1, 1])

v3a = v1+v2
v3b = np.add(v1,v2)
v3c = np.zeros(2)
for i in range(0,2):
    v3c[i] = v1[i] + v2[i]

print(v3a, v3b, v3c)
```

## showing what vector addition is like on a plot

```python
v1 = np.array([-3, 2])
v2 = np.array([2, 1])
v3 = v1 + v2

# plotting vectors in the standard position
plt.plot([0,v1[0]], [0, v1[1]], 'r', label="vector 1")
plt.plot([0,v2[0]]+v1[0], [0, v2[1]]+v1[1], 'b', label="vector 2")
plt.plot([0,v3[0]], [0, v3[1]], 'g', label="Combined Vector")

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```
