```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display,Math
```

```python
M = np.array([ 
    [.5,1], 
    [1,.5] ])

v = np.array([1,2])

Mv = M@v

plt.plot([0,v[0]], [0,v[1]],"b", linewidth=2, label="vector")
plt.plot([0,Mv[0]], [0,Mv[1]],"r--", linewidth=2, label="Mv")
plt.axis("square")
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
M = np.array([ 
    [.5,1], 
    [1,.5] ])

vcomp = np.linspace(-2,2,40)

# v[1]
for a in vcomp:
    v = np.array([1,a])
    Mv = M@v
    plt.plot([0,v[0]], [0,v[1]])
    plt.plot([0,Mv[0]], [0,Mv[1]])

plt.axis("square")
plt.axis("off")
plt.show()
```
