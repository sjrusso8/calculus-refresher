```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display,Math
```

## Vectors in Numpy

```python
row_vec = np.array([4,2])
col_vec = np.array([ [-2], [3]])
```

```python
display(Math(sym.latex(sym.sympify(row_vec))))
display(Math(sym.latex(sym.sympify(col_vec))))
```

```python
# plotting vectors in the standard position
plt.plot([0,row_vec[0]], [0, row_vec[1]], 'r', label="row vector")
plt.plot([0,col_vec[0]], [0, col_vec[1]], 'b', label="column vector")

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()

```

## Simple exercise for vectors

```python
vec1 = np.array([3,0,-4])
vec2 = np.array([-1,1,3])

display(Math(sym.latex(sym.sympify(vec1))))
display(Math(sym.latex(sym.sympify(vec2))))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=plt.figaspect(1))
# This is important
ax = fig.add_subplot(111, projection='3d')

# draw vectors
ax.plot([0,vec1[0]],[0,vec1[1]],[0,vec1[2]],'b',linewidth=3)
ax.plot([0,vec2[0]],[0,vec2[1]],[0,vec2[2]],'r',linewidth=3)

# guidelines
ax.plot([-5,5],[0,0],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[-5,5],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[0,0],[-5,5],'--',color=[.7,.7,.7])

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)
plt.show()
```
