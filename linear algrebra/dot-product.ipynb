## Dot Products

The output of a dot product is always a single number.  

![alt text](./images/dot-product.png "Image")



```python
# Creating two arrays of ranges
v = np.arange(10,20, 3)
w = np.arange(5, 15,3)

# Creating the dot product based on the big equation
dp1 = 0
for i in range(0, len(v)):
    dp1 = dp1 + v[i] * w[i]

# Using Numpy for multiplication
dp2 = np.sum( np.multiply(v,w))

# Using numpy dot
dp3 = np.dot(v,w)
print(dp1, dp2, dp3)
```

```python
v = np.arange(10,20, 3)
w = np.arange(5, 15,3)


def defineDotProduct(w,v):
    # Check if the two ranges are of the same length
    if (len(v) != len(w)):
        raise ValueError("Arrays are not of equal length")

    return np.dot(w,v)
        
```

## Application: Correlation CoEfficient

```python
# data
mikes_courses  = [4,5,8,2,9,9,1,3,7]
life_happiness = [6,7,9,3,9,3,1,6,7]

# mean-center
m = mikes_courses - np.mean(mikes_courses)
l = life_happiness - np.mean(life_happiness)

# compute correlation
numer = np.dot(m,l)
denom = np.sqrt( np.dot(m,m) ) * np.sqrt(np.dot(l,l))
r1 = numer/denom

# confirm with numpy function
r2 = np.corrcoef(m,l)[1][0]

# show the data!
plt.plot(mikes_courses,life_happiness,'ms',label='r=%s' %np.round(r1,2))

plt.axis([0,10,0,10])
plt.gca().set_aspect('equal')
plt.xlabel('Number of Mike\'s courses taken')
plt.ylabel('Life happiness')
plt.legend()
plt.show()
```

## Testing if these are commutative

a * b == b * a

Matrix 

np.outer(v,w) == np.outer(w,v)

Scalar Multiplication commutative

```python
v1 = np.arange(1,11)
v2 = np.arange(1,6)
print(np.outer(v1,v2))
print(np.outer(v2,v1))
np.outer(v1,v2) == np.outer(v2,v1)
```

```python
s = 4

res1 = s*np.outer(v1,v2)
res2 = np.outer(s*v1,v2)
res3 = np.outer(v1,s*v2)
res4 = np.outer(v1,v2)*s

print(res1 - res2)
print(res1 - res3)
print(res1 - res4)
```
