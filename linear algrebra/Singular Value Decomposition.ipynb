```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## Singular Value Decomposition

A(mxn) = U(mxm) * sigma(mxn) * V(nxn)^t

- U is the orthogonal basis for column space A. Orthogonal simply is 'perpendicular' and column space because it is m x m of the original size Amxn

- Sigma is the singular values of A in a diagonal matrix

- V transpose is the orthogonal basis of the ROW space of A.  Therefore, the size is n x n


U and V are identifying features of A as they are the column and row space respectively. Sigma is representing the importance of those features


```python
# import an image
ein = Image.open("./images/einstein.jpg")

# The matrix shape is the column, rows, and RGB color channel
# This a tensor / cube
print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
# Remove the gray scale
ein = np.mean(ein, 2)

print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)

print(np.shape(U)) # The same as the m x m
print(np.shape(s)) 
print(np.shape(V))

fig, ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title("U Matrix")
ax[1].imshow(np.diag(np.log(s))) # Add the log to scale down the values
ax[1].set_title("U Matrix")
ax[2].imshow(V)
ax[2].set_title("V Matrix")

plt.show()
```

```python
# Eigenspectrum

plt.plot(s, 'ks-', markerfacecolor="w")
plt.xlim([-1,50])
plt.title('Eigenspectrum of Ein')
plt.show()
```

## What does that first feature do?

```python
S = np.zeros(np.shape(ein))
for i in range(0,1):
    S[i,i] = s[i]

RevEin = U@S@V

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(RevEin)
plt.axis('off')
plt.title('SVD Version')

plt.show()
```
