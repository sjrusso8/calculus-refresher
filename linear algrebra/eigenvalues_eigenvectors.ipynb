## SECTION: Eigenvalues 



```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display,Math
```

```python
M = np.random.randint(-5,6,(5,5))

eigvalues, eigvecs = np.linalg.eig(M@M.T)

fig, ax = plt.subplots(1,3)

ax[0].imshow(M)
ax[0].set_title("Original Matrix")
ax[1].imshow(np.diag(eigvalues))
ax[1].set_title("Eigenvalues")
ax[2].imshow(eigvecs)
ax[2].set_title("Eigenvectors")
```
